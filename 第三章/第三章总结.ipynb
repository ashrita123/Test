```python
#5.文本情感分析
#通常步骤：
#一、读取data/dict中字典；二、读取data/txts中txt；三、定义计算情感的函数senti并计算分数；四、新建csv并存储；五、for循环
```

```python
#一、读取dict
poss=open('data/dict/positive.txt',encoding='gbk').read().split('\n')
negs=open('data/dict/negative.txt',encoding='gbk').read().split('\n')
```

```python
#二、读取txt
from pathlib import Path
txtdir=Path().cwd().joinpath('data','txts')#寻找文件路径
for file in txtdir.glob('*.txt'):
    file=str(file)
    text=open(file,encoding='gbk').read()
    raw=file.split('/')[-1][:-4]
    name=raw[-8:-4]
    year=raw[-4:]
    print(name,year,text[:10])
```

```python
#三、定义计算情感函数
import jieba
def senti(text):
    pos_num=0
    neg_num=0
    words=jieba.lcut(text)
    for word in words:
        if word in poss:
            pos_num=pos_num+1
        if word in negs:
            neg_num=neg_num+1
    return{"pos":pos_num,"neg":neg_num}

file='data/txts/联美控股2017.txt'
text=open(file,encoding='gbk').read()
senti(text)
```

```python
#四、新建csv
import csv
csvf=open('output/txts情感计算.csv','a+',encoding='utf-8')
fieldnames=['name','year','text','pos','neg']
writer=csv.DictWriter(csvf,fieldnames=fieldnames)
writer.writeheader()

file='data/txts/两面针2017.txt'
text=open(file,encoding='gbk').read()
raw=file.split('/')[-1][:-4]
name=raw[-8:-4]
year=raw[-4:]
score=senti(text)
data={'name':name, 'year':year, 'text':text, 'pos':score['pos'], 'neg':score['neg']}
writer.writerow(data)

csvf.close()
```

```python
#五、for循环
import jieba
from pathlib import Path
import csv

def senti(text):
    pos_num=0
    neg_num=0
    words=jieba.lcut(text)
    for word in words:
        if word in poss:
            pos_num=pos_num+1
        if word in negs:
            neg_num=neg_num+1
    return{"pos":pos_num,"neg":neg_num}

csvf=open('output/txts情感计算.csv','a+',encoding='utf-8')
fieldnames=['name','year','text','pos','neg']
writer=csv.DictWriter(csvf,fieldnames=fieldnames)
writer.writeheader()

poss=open('data/dict/positive.txt',encoding='gbk').read().split('\n')
negs=open('data/dict/negative.txt',encoding='gbk').read().split('\n')

txtdir=Path().cwd().joinpath('data','txts')#寻找文件路径
for file in txtdir.glob('*.txt'):
    file=str(file)
    text=open(file,encoding='gbk').read()
    raw=file.split('/')[-1][:-4]
    name=raw[-8:-4]
    year=raw[-4:]
    score=senti(text)
    data={'name':name, 'year':year, 'text':text, 'pos':score['pos'], 'neg':score['neg']}
    writer.writerow(data)

csvf.close()
```

```python
#step1
poss = open('data/dict/positive.txt',encoding='gbk').read().split('\n')
negs = open('data/dict/negative.txt',encoding='gbk').read().split('\n')

import jieba
def pos(text):
    num=0
    words=jieba.lcut(text)
    for word in words:
        if word in poss:
            num+=1
    return num

def neg(text):
    num=0
    words=jieba.lcut(text)
    for word in words:
        if word in negs:
            num+=1
    return num

text='XX'
pos(text)#neg(text)
```

```python
#step1.读取数据
import pandas as pd
df=pd.read_excel('data/small_mda_data.xlsx',converters={'股票代码':str})
df.head()
```

```python
#step2.构建词典
import cntext as ct
Chinese_Digitalization_Info=ct.load_pkl_dict('Chinese_Digitalization.pkl')
print(Chinese_Digitalization_Info)
```

```python
#查询Chinese_Digitalization下的关键title
Chinese_Digitalization_Info['Chinese_Digitalization'].keys()
```

```python
#step3.定义数字化函数
import cntext as ct
import pandas as pd

digtal_diction=ct.load_pkl_dict('Chinese_Digitalization.pkl')['Chinese_Digitalization']
def digtal_function(text):
    res=ct.sentiment(text=text,diction=digtal_diction)
    return pd.Series(res)

test_text='近年来，我国在大数据、人工智能、云计算、工业互联网等领域取得了重大进展，部分先进技术已具备产业化基础。'
digtal_function(text=test_text)
```

```python
#step4.批量计算
res_df=df['经营讨论与分析内容'].apply(digtal_function)
res_df.head()
```

```python
#step5.结果整理
res_df['Digital_word_num']=res_df[['Artificial_Intelligence_num','Big_Data_num',
                                  'Cloud_Computing_num','Block_Chains_num',
                                  'Usage_of_Digitalization_num']].sum(axis=1)
res_df['Digital_Index']=res_df['Digital_word_num']/res_df['word_num']
res_df.head()
```

```python
#step6.保存结果
df2=pd.concat([df,res_df],axis=1)
df2.head()
```
